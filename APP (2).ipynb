```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

accidents = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Thư mục chứa file CĐ/Accident_Information.csv')
accidents['Date'] = pd.to_datetime(accidents['Date'])
accidents['Year'] = accidents['Date'].dt.year
accidents['Month'] = accidents['Date'].dt.month
accidents['Day'] = accidents['Date'].dt.day

# Phân tích số lượng tai nạn theo năm
accidents_by_year = accidents['Year'].value_counts().sort_index()
print(accidents_by_year)
```

```python
# Phân phối các yếu tố ảnh hưởng
sns.countplot(x='Weather_Conditions', data=accidents)
plt.xticks(rotation=45)
plt.show()
```

```python
# Tác động của loại đường và điều kiện đường đến mức độ nghiêm trọng của tai nạn
sns.countplot(x='Road_Surface_Conditions', hue='Accident_Severity', data=accidents)
plt.xticks(rotation=45)
plt.show()
```

```python
# Tác động của thời tiết và ánh sáng đến tai nạn
sns.countplot(x='Light_Conditions', hue='Weather_Conditions', data=accidents)
plt.xticks(rotation=45)
plt.show()
```

```python
# Tương quan giữa các biến số
correlation_matrix = accidents.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Mối quan hệ giữa tốc độ và mức độ nghiêm trọng của tai nạn
sns.boxplot(x='Accident_Severity', y='Speed_limit', data=accidents)
plt.show()
```
